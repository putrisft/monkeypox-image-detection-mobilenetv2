# Augment Data

```python
datagen = ImageDataGenerator(
            rotation_range=45,
            width_shift_range=0.2,
            height_shift_range=0.2,
            shear_range=0.2,
            zoom_range=0.2,
            horizontal_flip=True,
            fill_mode='reflect')
```

```python
i = 0
for batch in datagen.flow_from_directory(directory='drive/MyDrive/DATA/skripsi/dataset-converted/',
                                         batch_size=150,
                                         target_size=(128,128),
                                         color_mode='rgb',
                                         save_to_dir='drive/MyDrive/DATA/skripsi/new/',
                                         save_prefix='aug',
                                         save_format='png'):
  i+=1
  if i > 19: #Epochs
    break
```

# Visualize

```python
import glob
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
%matplotlib inline

images = []
for img_path in glob.glob('drive/MyDrive/DATA/skripsi/dataset-augment-all-sample/*.png'):
    images.append(mpimg.imread(img_path))

plt.figure(figsize=(10,10))
for i, image in enumerate(images):
    plt.subplot(4, 7, i + 1)
    plt.imshow(image)
    plt.xticks([])
    plt.yticks([])
```
